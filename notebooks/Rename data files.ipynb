```python
from pathlib import Path

INPUT_SRC = "../data/phase_two/subclonal_structure/"
```

Resolve and create a temporary location for renamed data files

```python
input_path = Path(INPUT_SRC)
temp_folder = "." + input_path.parts[-1]  # Make a hidden folder with the same name 
output_path = input_path.parent.joinpath(temp_folder)

if not output_path.exists():
    output_path.mkdir()

print(output_path)
```

Rename files to include just the original filename and `.tsv` extension

```python
for data_file in input_path.iterdir():
    new_filename = data_file.stem.split(".")[0] + ".tsv"
    print(data_file.rename(output_path.joinpath(new_filename)))
```

Delete the (now) empty `subclone_stucture` directory and rename the temporary `.subclonal_structure` directory to `subclonal_structure`

```python
output_path.replace(input_path)
```
